#### For the final project, I decided to use the Parkinson and tele-monitoring dataset. This dataset has 5875 rows, no missing values, and 22 columns of which 20 are features. The columns motor_UPDRS and total_UPDRS are the response based on the data. Those values are something called UPDRS score which measures the severity of Parkinson's disease in patients, with 0 being no disease to 199 being the worst. For this dataset, regression is available. Therefore, I will do a regression problem where the motor_UPDRS total_UPDRS response will be predicted based on the features.

```python
import pandas as pd
df = pd.read_csv("parkinsons.csv")
df
```

## Step Two
In this step, you must identify the relevant columns in the dataset that can be used as predictors. Any irrelevant data columns must be excluded from the analysis. If there are any missing values in the dataset, they should be replaced through a data imputation method. Any missing rows should be deleted from the dataset. Each column data type must be converted into an appropriate data type. The data types can be obtained either in data info files or can be determined by inspection.

```python
df.dropna(how = "all", inplace = True)
df = df.drop(columns = ["subject#", "age", "sex", "test_time"])
# all varibles are continuous, therefore we will make sure they are converted to float
for i in df.columns.tolist():
    df[i] = df[i].apply(lambda x: float(x))
df
```

## Step Five
This step should start with creating two dataframes: response and features. Split these dataframes into train and test parts.


```python
from sklearn.model_selection import train_test_split
features = df.drop(columns = ["motor_UPDRS", "total_UPDRS"], inplace = False)
response = df[["motor_UPDRS", "total_UPDRS"]]
features_train, features_test, response_train, response_test = train_test_split(features, response, test_size = 0.2, random_state = 0)
```

## Step Six
Choose an appropriate model that can be used for your problem. You can choose a model that is listed under a given problem group below. You are also free to choose another model that you are familiar with. Fit your train data to your model. Generate predictions over the test data and then evaluate your model by reporting appropriate accuracy metrics.

```python
# we will do regression to predict motor_UPDRS and total_UPDRS values
# we will 
import pandas as pd
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn import metrics
import numpy as np

models = [DecisionTreeRegressor(), RandomForestRegressor(), LinearRegression()]
models_names = ["Decision Tree", "Random Forest", "Linear Regression"]
data = []

for i in range(len(models)):
    model = models[i]
    model.fit(features_train, response_train)
    response_pred = model.predict(features_test)
    temp = []
    temp.append(models_names[i])
    temp.append(metrics.mean_absolute_error(y_true = response_test, y_pred = response_pred))
    temp.append(metrics.mean_squared_error(y_true = response_test, y_pred = response_pred))
    temp.append(np.sqrt(metrics.mean_squared_error(y_true = response_test, y_pred = response_pred)))
    data.append(temp)
    temp = []

the_names = ["Model", "Mean Absolute Error", "Mean Square Error", "Root Mean Square Error"]
results = pd.DataFrame(data, index = None, columns = the_names)

results

#random forest regressor yielded the lowest error on all metrics. therefore, it will be used for the rest of the questions
```

## Step Seven
You must report the feature importance results.

```python
forest = RandomForestRegressor(random_state = 0)
forest.fit(features_train, response_train)
response_pred = forest.predict(features_test)
```

```python
# first 200 rows were selected because sample is too big
import matplotlib.pyplot as plot
plot.figure(figsize = (7,6))
plot.scatter(response_test.iloc[0:200], response_pred[0:200, :]) 
# we only used the first 200 rows becuase the sample is too big
plot.xlabel("Actual", size = 12)
plot.ylabel("Prediction", size = 12)
plot
```

```python
imp = forest.feature_importances_.tolist()
feature_names = features.columns.tolist()
temp = []
for i in range(len(imp)):
    temp.append([feature_names[i], imp[i]])
importance = pd.DataFrame(temp, columns = ["Feature", "Importance_score"])
importance
```

```python
# the error metircs reported above were pretty high. Therefore, the data will be normalized and then the best model will 
# be selected for further analysis. Also, some columns will be dropped

df_clean = df.drop(columns = ["Jitter:RAP", "Jitter:DDP", "Shimmer", "Shimmer:DDA"], inplace = False)

from sklearn.preprocessing import MinMaxScaler
data_nor = MinMaxScaler().fit_transform(df_clean) 
df_nor = pd.DataFrame(data_nor, columns = df_clean.columns.tolist())
df_nor
```

```python
from sklearn.model_selection import train_test_split
features = df_nor.drop(columns = ["motor_UPDRS", "total_UPDRS"], inplace = False)
response = df_nor[["motor_UPDRS", "total_UPDRS"]]
features_train, features_test, response_train, response_test = train_test_split(features, response, test_size = 0.2, random_state = 0)
```

```python
# we will do regression to predict motor_UPDRS and total_UPDRS values
# we will 
import pandas as pd
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn import metrics
import numpy as np

models = [DecisionTreeRegressor(), RandomForestRegressor(), LinearRegression()]
models_names = ["Decision Tree", "Random Forest", "Linear Regression"]
data = []

for i in range(len(models)):
    model = models[i]
    model.fit(features_train, response_train)
    response_pred = model.predict(features_test)
    temp = []
    temp.append(models_names[i])
    temp.append(metrics.mean_absolute_error(y_true = response_test, y_pred = response_pred))
    temp.append(metrics.mean_squared_error(y_true = response_test, y_pred = response_pred))
    temp.append(np.sqrt(metrics.mean_squared_error(y_true = response_test, y_pred = response_pred)))
    data.append(temp)
    temp = []

the_names = ["Model", "Mean Absolute Error", "Mean Square Error", "Root Mean Square Error"]
results = pd.DataFrame(data, index = None, columns = the_names)

results

# random fores regressor yielded the lowest error on all metrics. therefore, it will be used for the rest of the questions
# we can see that dropping some features and normalizing the data yields better results. 
```
